### Line Graph

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx", engine='openpyxl')
df.head()
```

```python
df.tail()
```

```python
# Check for NaN
df.isnull().sum()
```

```python
# Fill missing values (NaN) with front data
df.fillna(method="ffill", inplace=True)
```

```python
df
```

```python
# Check for NaN
df.isnull().sum()
```

```python
# Extract and organize data of migration from Seoul to other regions
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시') 
mask
```

```python
df_seoul = df[mask]
```

```python
df_seoul.head()
```

```python
# Delete columns by place of origin
df_seoul=df_seoul.drop(['전출지별'], axis=1)
```

```python
df_seoul.head()
```

```python
# '전입지별'을 '전입지'로 컬럼 이름 변경

df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.head()
```

```python
# Change the index of df_seoul to '전입지'
df_seoul.set_index('전입지', inplace=True)
df_seoul.head()
```

```python
# Select only the population migration data from Seoul to Gyeonggi
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
# Draw a line graph
plt.plot(sr_one.index, sr_one.values)
```

```python
plt.plot(sr_one)
```

```python
# Solve Korean font issue

# Matplotlib does not support Korean fonts
# OS information
import platform

# font_manager: Font management module
# rc: Font change module
from matplotlib import font_manager, rc
# Unicode settings
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# Add chart title
plt.plot(sr_one)
plt.title('서울 -> 경기 인구 이동')

# Add axis labels
plt.xlabel('기간')
plt.ylabel('이동 인구수')

plt.show() 
```

# 동일한 그림에 여러개의 그래프를 추가

```python
# Select only the population migration data from Seoul to 'Chungcheongnam-do', 'Gyeongsangbuk-do', and 'Gangwon-do'
df_seoul.head()
```

```python
col_years = list(map(str,range(1970, 2018)))
df_3 = df_seoul.loc[['충청남도', '경상북도','강원도'], col_years]
df_3
```

```python
# Style
plt.style.use('ggplot')

# Create graph object
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1,1,1)

# Output data
ax.plot(col_years, 
        df_3.loc['충청남도'],
        marker='o',
        markerfacecolor='green',
        markersize=10,
        color='green',
        linewidth=2,
        label='서울->충남'
       )

ax.plot(col_years, 
        df_3.loc['경상북도'],
        marker='o',
        markerfacecolor='blue',
        markersize=10,
        color='skyblue',
        linewidth=2,
        label='서울->경북'
       )

# Gangwon-do
ax.plot(col_years, 
        df_3.loc['강원도'],
        marker='o',
        markerfacecolor='red',
        markersize=10,
        color='magenta',
        linewidth=2,
        label='서울->강원도'
       )

# Chart title: Migration from Seoul to Chungnam, Gyeongbuk, Gangwon, size: 20
ax.set_title('서울->충남,경북, 강원 인구 이동', size=20)

# Add axis labels: X -> Period, Y -> Migrating Population, size: 12
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)

# Show legend
ax.legend(loc='best')

# Specify x-axis tick labels and rotate 90 degrees
ax.set_xticklabels(col_years, rotation=90)

# Tick label size
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```

> 서울 특별시에서 특히 지리적으로 가까운 충남 지역으로 이동한 인구가 다른 두지역에 비해 많은 편이다.
> 전반적으로 1970~1980년대에는 서울에서 지방으로 전출하는 인구가 많았으나 1990년 이후 줄곧 감소하는 패턴을 보이고 있다.

```python
# Extract only the population migration data from Seoul to 'Chungcheongnam-do', 'Gyeongsangbuk-do', 'Gangwon-do', and 'Jeollanam-do'
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
```

```python

plt.style.use('ggplot')


fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(2,2,1)


ax1.plot(col_years,
         df_4.loc['충청남도'],
         marker = 'd',
         markerfacecolor = 'green',
         markersize = 10,
         color='olive',
         linewidth = 2,
         label = '서울 -> 충남'
        )


ax1.legend(loc='best')


ax1.set_title('서울->충남 인구 이동', size=15)


ax1.set_xticklabels(col_years, rotation=90, size=6)

plt.savefig("../Data/seoul.jpg")
plt.show() 
```

# Area Graph (Stacked Line Graph)

### Overlapping Graph of Cumulative Data

```python
df_4 = df_4.T
df_4
```

```python

plt.style.use('ggplot')

df_4.plot(kind='area',
          stacked=False,
          alpha=0.2, 
          figsize=(20, 10) 
         )

plt.title('서울->타시도 인구이동', size=30)
plt.ylabel('이동 인구수', size=20)
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

### Non-overlapping Graph of Cumulative Data

```python

plt.style.use('ggplot')

df_4.plot(kind='area',
          stacked=True, 
          alpha=0.2, 
          figsize=(20, 10) 
         )

plt.title('서울->타시도 인구이동', size=30, color='brown', weight='bold')
plt.ylabel('이동 인구수', size=20, color='blue')
plt.xlabel('기간', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

### Comparison of Migration In and Out of Seoul and Busan with a Line Graph

```python
# Seoul -> Busan
sr_seoul = df_seoul.loc['부산광역시']
sr_seoul.head()
```

```python
# Busan-> Seoul
sr_busan = df_busan.loc['서울특별시']
sr_busan.head()
```

```python

plt.style.use('ggplot')


fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1,1,1)

ax.plot(col_years,
        sr_seoul,
        marker='o',
        markerfacecolor='red',
        markersize=10,
        color='red',
        linewidth=2,
        label='서울->부산'
       )

ax.plot(col_years,
        sr_busan,
        marker='o',
        markerfacecolor='blue',
        markersize=10,
        color='skyblue',
        linewidth=2,
        label='부산->서울'
       )


ax.legend(loc='best')


ax.set_title('서울<->부산 인구 이동', size=20)

ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)

ax.set_xticklabels(col_years, rotation=90)


ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```

# Horizontal Bar Graph Compare the total population migrated from Seoul to each province from 2010 to 2017 using a graph

```python
col_years = list(map(str,range(2010, 2018)))
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'], col_years]
df_4
```

```python
# Add a new column by summing the migration population from 2010 to 2017

df_4['합계'] = df_4.sum(axis=1)
df_4.head()
```

```python
# Sort by the largest sum
# Create graphs in ascending order by the first row of data

df_total = df_4[['합계']].sort_values(by='합계', ascending=True)
df_total
```

```python

plt.style.use('ggplot')

# Horizontal bar graph

df_total.plot(kind='barh',
                  color = 'skyblue',
                  width=0.5,
                  figsize=(10,5)
               )

plt.title('서울->타시도 인구 이동')
plt.xlabel('이동 인구수')
plt.ylabel('전입지')

plt.show()
```
